```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

# ── SECTION 2: Delivery Analysis ───────────────────────────────

# Chart 1: Late vs On-Time
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

delivery_counts = data['delivery_status'].value_counts()
axes[0].bar(delivery_counts.index, delivery_counts.values,
            color=['#e74c3c','#2ecc71','#3498db','#f39c12'])
axes[0].set_title('Order Count by Delivery Status', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Delivery Status')
axes[0].set_ylabel('Number of Orders')
axes[0].tick_params(axis='x', rotation=15)

# Chart 2: Late delivery rate by shipping mode
late_rate = data.groupby('shipping_mode')['late_delivery_risk'].mean() * 100
axes[1].bar(late_rate.index, late_rate.values, color='#e74c3c')
axes[1].set_title('Late Delivery Rate by Shipping Mode (%)', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Shipping Mode')
axes[1].set_ylabel('Late Delivery Rate (%)')
axes[1].tick_params(axis='x', rotation=15)

plt.tight_layout()
plt.savefig('delivery_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("Insight: Standard Class has the highest late delivery rate despite being the most used mode.")
```

```python
# ── SECTION 3: Customer Analysis ───────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Chart 3: Customer count by segment
seg_counts = data['customer_segment'].value_counts()
axes[0].bar(seg_counts.index, seg_counts.values,
            color=['#3498db','#e74c3c','#2ecc71'])
axes[0].set_title('Customer Count by Segment', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Customer Segment')
axes[0].set_ylabel('Count')

# Chart 4: Revenue by segment
seg_revenue = data.groupby('customer_segment')['sales'].sum().sort_values(ascending=False)
axes[1].bar(seg_revenue.index, seg_revenue.values,
            color=['#3498db','#e74c3c','#2ecc71'])
axes[1].set_title('Total Revenue by Customer Segment', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Customer Segment')
axes[1].set_ylabel('Total Revenue ($)')
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))

plt.tight_layout()
plt.savefig('customer_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("Insight: Consumer segment dominates both order volume and total revenue.")
```

```python
# ── SECTION 4: Product Profitability ───────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Chart 5: Top 10 products by revenue
top_products = data.groupby('product_name')['sales'].sum().nlargest(10)
axes[0].barh(top_products.index, top_products.values, color='#3498db')
axes[0].set_title('Top 10 Products by Revenue', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Total Revenue ($)')
axes[0].invert_yaxis()

# Chart 6: Discount rate vs profit ratio scatter
sample = data.sample(2000, random_state=42)
axes[1].scatter(sample['order_item_discount_rate'],
                sample['order_item_profit_ratio'],
                alpha=0.3, color='#e74c3c', s=20)
axes[1].axhline(y=0, color='black', linestyle='--', linewidth=1)
axes[1].set_title('Discount Rate vs Profit Ratio', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Discount Rate')
axes[1].set_ylabel('Profit Ratio')

plt.tight_layout()
plt.savefig('product_profitability.png', dpi=150, bbox_inches='tight')
plt.show()
print("Insight: Higher discount rates strongly correlate with negative profit ratios.")
```

```python
# ── SECTION 5: Regional Analysis ───────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Chart 7: Revenue by market
market_rev = data.groupby('market')['sales'].sum().sort_values(ascending=False)
axes[0].bar(market_rev.index, market_rev.values, color='#9b59b6')
axes[0].set_title('Total Revenue by Market', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Market')
axes[0].set_ylabel('Total Revenue ($)')
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
axes[0].tick_params(axis='x', rotation=15)

# Chart 8: Late delivery rate by market
market_late = data.groupby('market')['late_delivery_risk'].mean() * 100
axes[1].bar(market_late.index, market_late.values, color='#e67e22')
axes[1].set_title('Late Delivery Rate by Market (%)', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Market')
axes[1].set_ylabel('Late Delivery Rate (%)')
axes[1].tick_params(axis='x', rotation=15)

plt.tight_layout()
plt.savefig('regional_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("Insight: LATAM has highest revenue but also one of the highest late delivery rates.")
```

```python
# ── SECTION 6: Time Trend ───────────────────────────────────────
data['order_date'] = pd.to_datetime(data['order_date_dateorders'],
                                     format='%m/%d/%Y %H:%M')
data['year_month'] = data['order_date'].dt.to_period('M')

monthly = data.groupby('year_month').agg(
    total_orders=('order_id', 'count'),
    total_revenue=('sales', 'sum'),
    late_rate=('late_delivery_risk', 'mean')
).reset_index()

fig, axes = plt.subplots(2, 1, figsize=(14, 10))

# Chart 9: Monthly order trend
axes[0].plot(monthly['year_month'].astype(str),
             monthly['total_orders'], color='#3498db', linewidth=2, marker='o', markersize=4)
axes[0].set_title('Monthly Order Volume (2015-2018)', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Number of Orders')
axes[0].tick_params(axis='x', rotation=45)
step = max(1, len(monthly) // 12)
axes[0].set_xticks(range(0, len(monthly), step))

# Chart 10: Monthly late delivery rate trend
axes[1].plot(monthly['year_month'].astype(str),
             monthly['late_rate'] * 100, color='#e74c3c', linewidth=2, marker='o', markersize=4)
axes[1].set_title('Monthly Late Delivery Rate % (2015-2018)', fontsize=13, fontweight='bold')
axes[1].set_ylabel('Late Delivery Rate (%)')
axes[1].set_xlabel('Month')
axes[1].tick_params(axis='x', rotation=45)
axes[1].set_xticks(range(0, len(monthly), step))

plt.tight_layout()
plt.savefig('time_trends.png', dpi=150, bbox_inches='tight')
plt.show()
print("Insight: Order volume grew steadily from 2015-2017 before dropping in 2018.")
```
